# merge()

```python
data4 = {'emp_id' : [101,102,103,104],
        'salary' : [654,432,200,433],
        'pf' : [32,45,23,23]}
```

```python
data5 = {'emp_id' : [101,102,104,105],
        'mob_no' : [22132,34322,45673,12312],
        'house_nu' : [345,345,345,56]}
```

```python
df4 = pd.DataFrame(data4)
```

```python
df4
```

```python
df5 = pd.DataFrame(data5)
```

```python
df5
```

```python
pd.merge(df4,df5) #bydefault find comman values, whatever comman data we have W.R.T comman columns are there
```

```python
pd.merge(df4,df5, how='right', on  = 'emp_id')
```

```python
data6 = {'emp_id1' : [101,102,103,104],
        'salary' : [234, 345, 564, 400],
        'pf' : [23,34,56,12]}

data7 = {'emp_id2' : [101,102,104,105],
        'mob_no' : [1212,3434,4545,8787],
        'house_no' : [232,343,565,787]
        }
#if comman col not there then it not merge
```

```python
df6 = pd.DataFrame(data6)
```

```python
df7 = pd.DataFrame(data7)
```

```python
pd.merge(df6,df7, left_on='emp_id1', right_on='emp_id2', how = 'inner') # inner bydefault
```
